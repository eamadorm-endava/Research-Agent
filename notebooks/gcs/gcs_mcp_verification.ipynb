This notebook demonstrates how to use the `GCSManager` and `build_gcs_credentials` from the GCS MCP server package. It follows the pattern of using delegated OAuth tokens for storage operations.

```python
import sys
import os
from loguru import logger

# Ensure the repository root is in sys.path
sys.path.append('../..')

from mcp_servers.gcs.app.gcs_client import GCSManager, build_gcs_credentials
from mcp_servers.gcs.app.config import GCS_API_CONFIG
from google.oauth2.credentials import Credentials
```

## Authentication Mockup

In a real scenario, the `access_token` would come from the user's OAuth flow or MCP context. For demonstration, we'll show how the credentials object is built.

```python
mock_access_token = "ya29.mock-token"
try:
    # Note: validate=True would attempt to hit Google's tokeninfo endpoint
    creds = build_gcs_credentials(access_token=mock_access_token, validate=False)
    print("Successfully built GCS credentials from token.")
except Exception as e:
    print(f"Failed to build credentials: {e}")
```

## Initializing GCSManager

The `GCSManager` orchestrates all interactions with the Cloud Storage API.

```python
# Mocking a project ID if not detected automatically
project_id = "your-gcp-project-id"

try:
    gcs_manager = GCSManager(creds=creds, default_project=project_id)
    print(f"GCSManager initialized for project: {project_id}")
except Exception as e:
    print(f"Initialization error (expected if mock creds are used): {e}")
```
