# Time Tool Verification

This notebook demonstrates the functionality of the `get_current_time` tool.

```python
import sys
import os
from datetime import datetime

# Add repository root to path
sys.path.append("../..")

from agent.core_agent.tools.time_tools import GetCurrentTimeTool
import asyncio
```

## Execute the Tool

We will run the `get_current_time` tool and verify the output.

```python
async def verify_tool():
    tool = GetCurrentTimeTool()
    result = await tool.run_async(args={}, tool_context=None)
    print(f"Execution Status: {result['execution_status']}")
    print(f"Current Time (ISO): {result['current_time']}")
    print(f"Timezone: {result['timezone']}")
    
    return result

# In Jupyter, we can use await directly or run it in an event loop
result = await verify_tool()
```

## Validation

Verify that the timezone offset is correct for Central Time (CST/CDT).

```python
dt = datetime.fromisoformat(result['current_time'])
offset_hours = dt.utcoffset().total_seconds() / 3600
print(f"Timezone Offset: {offset_hours} hours")

if offset_hours in [-6.0, -5.0]:
    print("✅ Success: Correct offset for Central Time.")
else:
    print("❌ Failure: Unexpected offset for Central Time.")
```
